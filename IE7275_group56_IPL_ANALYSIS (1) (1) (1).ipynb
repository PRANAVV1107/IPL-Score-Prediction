```python
# the overs format is 0.0 so we need to convert the string into float
df['overs'] = df['ball']
df['ball'] = df['ball'].astype(str)
df['over'] = df['ball'].str[0:-1]
df['over'] = df['over'].str.replace('.','')
df['ball'] = df['ball'].str[-1:]
```

```python
df
```

```python
#this feature is to compute the total runs scored which is the runs_off_bat + extras such as wides, noballs etc.
df['total_runs'] = df.apply(lambda x:x['runs_off_bat'] + x['extras'],axis=1)
df
```

```python
#this is to make a boolean value for wicket, 0 if it is not out, 1 if it is out
df['wicket'] = df['player_dismissed'].apply(lambda x: 1 if type(x) == type('str') else 0)
```

```python
df
```

```python
#filtering out the relevant columns
df = df[['match_id','innings','batting_team','bowling_team','overs','over','ball','total_runs','player_dismissed','season']]
df
```

```python
#renaming columns
df = df.rename(columns = {'match_id':'id'})
df
```

### Feature Engineering and Selection

```python
#creating a boolean for wickets taken
df['wicket'] = df['player_dismissed'].apply(lambda x: 1 if type(x) == type('str') else 0)
```

```python
df
```

```python
#replacing NaN values with 0 
df = df.replace(np.nan, 0)
```

```python
main_teams = ['Kolkata Knight Riders', 'Royal Challengers Bangalore',
           'Chennai Super Kings', 'Kings XI Punjab', 'Rajasthan Royals',
           'Delhi Daredevils', 'Deccan Chargers', 'Mumbai Indians',
           'Sunrisers Hyderabad', 'Delhi Capitals', 'Punjab Kings',
           ]
```

```python
#filtering out for main teams in the IPL that have played at least 5 seasons, batting and bowling teams
ndf = df[df['batting_team'].isin(main_teams)]
```

```python
ndf
```

```python
ndf = df[df['bowling_team'].isin(main_teams)]
ndf
```

```python
ndf
```

```python
ndf.groupby(['id','innings'])['total_runs'].transform('sum')
```

```python
#to find the total runs at the end of each innings
ndf['total'] = ndf.groupby(['id','innings'])['total_runs'].transform('sum')
ndf
```

```python
#to find the cumulative score in order to derive deeper features
ndf['rolling_score'] = ndf.groupby(['id','innings'])['total_runs'].apply(lambda x: np.cumsum(x))
ndf
```

```python
#the code lines below runs scored, wickets taken and boundaries in the previous 18 balls i.e. a 3 over window per innings
prev_18_balls = ndf.groupby(['id','innings'])['total_runs'].rolling(min_periods=1, window = 18).sum().reset_index()
prev_18_balls
```

```python
# score in last 3 overs
ndf['score_3_overs'] = prev_18_balls['total_runs'].tolist()
ndf
```

```python
ndf['player_dismissed'] = np.where(ndf['player_dismissed'] == 0,0,1) 
```

```python
ndf
```

```python
#wickets in last 3 overs
prev_3_wicks = ndf.groupby(['id','innings'])['player_dismissed'].rolling(min_periods=1, window = 18).sum().reset_index()
ndf['wickets_3_overs']  =  prev_3_wicks['player_dismissed'].to_list()
ndf.head(100)
```

```python
ndf['total_wickets'] = ndf.groupby(['id','innings'])['player_dismissed'].apply(lambda x : np.cumsum(x))
```

```python
ndf
```

```python
#dot balls (0s) in the last 3 overs
ndf['dot_balls'] = ndf['total_runs']
ndf['dot_balls'] = np.where(ndf['dot_balls']==0,1,0)
dots = ndf.groupby(['id','innings'])['dot_balls'].rolling(min_periods=1, window = 18).sum().reset_index()
ndf['dot_balls'] = dots['dot_balls'].tolist()
ndf
```

```python
#boundaries in the last 3 overs
ndf['boundaries'] = ndf['total_runs']
ndf['boundaries'] = np.where(ndf['boundaries']>3,1,0)
dots = ndf.groupby(['id','innings'])['boundaries'].rolling(min_periods=1, window = 18).sum().reset_index()
ndf['boundaries'] = dots['boundaries'].tolist()
ndf
```

### Function to compare model metrics

```python
#writing a function to evaluate performance of the models taken
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def evaluate_regression_models(X_train, X_test, y_train, y_test):
    # Initialize the models
    lr_model = LinearRegression()
    rf_model = RandomForestRegressor(n_estimators=100)
    gb_model = GradientBoostingRegressor(n_estimators=100)
    lasso_model = Lasso(alpha=0.1, random_state=42)
    # Fit the models on the training data
    lr_model.fit(X_train, y_train)
    rf_model.fit(X_train, y_train)
    gb_model.fit(X_train, y_train)
    lasso_model.fit(X_train,y_train)

    # Make predictions on the testing data
    y_pred_lr = lr_model.predict(X_test)
    y_pred_rf = rf_model.predict(X_test)
    y_pred_gb = gb_model.predict(X_test)
    y_pred_lasso = lasso_model.predict(X_test)
    # Calculate and print the evaluation metrics
    mse_lr = mean_squared_error(y_test, y_pred_lr)
    mse_rf = mean_squared_error(y_test, y_pred_rf)
    mse_gb = mean_squared_error(y_test, y_pred_gb)
    mse_lasso = mean_squared_error(y_test, y_pred_lasso)
    print("MSE - Linear Regression: {:.2f}, Random Forest: {:.2f}, Gradient Boosting: {:.2f}, Lasso Regression: {:.2f}".format(mse_lr, mse_rf, mse_gb,mse_lasso))

    mae_lr = mean_absolute_error(y_test, y_pred_lr)
    mae_rf = mean_absolute_error(y_test, y_pred_rf)
    mae_gb = mean_absolute_error(y_test, y_pred_gb)
    mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
    print("MAE - Linear Regression: {:.2f}, Random Forest: {:.2f}, Gradient Boosting: {:.2f}, Lasso Regression: {:.2f}".format(mae_lr, mae_rf, mae_gb,mae_lasso))

    r2_lr = r2_score(y_test, y_pred_lr)
    r2_rf = r2_score(y_test, y_pred_rf)
    r2_gb = r2_score(y_test, y_pred_gb)
    r2_lasso = r2_score(y_test, y_pred_lasso)
    print("R-squared - Linear Regression: {:.2f}, Random Forest: {:.2f}, Gradient Boosting: {:.2f}, Lasso Regression: {:.2f} ".format(r2_lr, r2_rf, r2_gb,r2_lasso))
```
